```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

models = ['gpt-4o-2024-05-13', 'claude-3-5-sonnet-20240620', 
          'claude-3-opus-20240229', 'Mistral-7B-T', 
          'llama2-7B-Q8', 'Llama-2-13b']

plt.style.use('seaborn-v0_8-paper')

def clean_data(df, model_col):
    df_filtered = df.copy()
    df_filtered[model_col] = pd.to_numeric(df_filtered[model_col], errors='coerce')
    df_filtered = df_filtered.dropna(subset=[model_col])
    return df_filtered

category_order = ['Short', 'Medium', 'Long']

for model in models:
    model_col = f'{model}_confidence-num'
    df_clean = clean_data(df_merged, model_col)
    
    df_clean['Qtype Length'] = pd.Categorical(df_clean['Qtype Length'], categories=category_order, ordered=True)

    plt.figure(figsize=(10, 8))
    ax = sns.boxplot(x='Qtype Length', y=model_col, data=df_clean,
                     order=category_order,
                     fliersize=3)
    ax.set_facecolor('#f8f8f8')
    plt.title(f'{name_mapping[model]}', fontsize=16, weight='bold')
    plt.ylabel('Confidence Number', fontsize=14)
    plt.xlabel('Qtype Length', fontsize=14)
    
    plt.ylim(0, 10)
    plt.yticks(fontsize=12)
    plt.xticks(fontsize=12)
    
    plt.grid(True, linestyle='--', alpha=0.7)

    # Annotate each box plot with detailed statistics
    for i, category in enumerate(category_order):
        data = df_clean[df_clean['Qtype Length'] == category][model_col]
        count_val = data.count()
        median_val = data.median()
        q1 = data.quantile(0.25)
        q3 = data.quantile(0.75)
        
        # Detailed annotation including median, Q1, Q3, and count
        annotation_text = (
            f"n={count_val}\n"
            f"Median={median_val:.2f}\n"
            f"Q1={q1:.2f}, Q3={q3:.2f}"
        )
        
        # Annotate with a dark blue color on a light gray background for readability
        ax.annotate(annotation_text,
                    xy=(i, 0.1), 
                    xytext=(0, 0), 
                    textcoords='offset points',
                    ha='center',
                    fontsize=13, 
                    fontweight='bold',
                    color='#032B44',
                    bbox=dict(boxstyle="round", alpha=0.8, fc='#F2F2F2'))

    # Adjust the y-axis limits to accommodate the annotations
    ax.set_ylim(0, 10)
    plt.gcf().tight_layout()

    plt.savefig(f'boxplt/confidence_boxplot_{model}.png', dpi=300, bbox_inches='tight')
    plt.savefig(f'boxplt/confidence_boxplot_{model}.pdf', dpi=300, bbox_inches='tight')
    plt.savefig(f'boxplt/confidence_boxplot_{model}.jpg', dpi=300, bbox_inches='tight')

    
    plt.show()
```

```python
from PIL import Image
import os

def stack_images_horizontally(image_folder, output_path, spacing=10):
    # Get all PNG files in the folder
    image_paths = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.lower().endswith('.png')]
    
    # Sort the image paths to ensure consistent ordering
    image_paths.sort()
    
    # Open all images
    images = [Image.open(path) for path in image_paths]
    
    # Calculate the total width and maximum height
    total_width = sum(img.width for img in images) + spacing * (len(images) - 1)
    max_height = max(img.height for img in images)
    
    # Create a new image with the calculated dimensions
    new_image = Image.new('RGB', (total_width, max_height), color='white')
    
    # Paste images horizontally
    x_offset = 0
    for img in images:
        new_image.paste(img, (x_offset, (max_height - img.height) // 2))
        x_offset += img.width + spacing
    
    # Save the result
    new_image.save(output_path, 'PNG', dpi=(300, 300))  # Set DPI to 300 for high resolution

    return len(images)  # Return the number of images processed

# Example usage
listt = ['boxplt/bad','boxplt/good']
for item in listt:
    image_folder = item
    output_path = f'./{image_folder}.png'

    num_images = stack_images_horizontally(image_folder, output_path)
    print(f"Stacked {num_images} images. Result saved to {output_path}")
```
