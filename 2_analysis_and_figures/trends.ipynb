```python
import pandas as pd
import matplotlib.pyplot as plt

models = ['gpt-4o-2024-05-13', 'claude-3-5-sonnet-20240620','claude-3-opus-20240229','Mistral-7B-T','llama2-7B-Q8','Llama-2-13b']
window_size = 2  # Adjust this window size based on how smooth you want the trend to be
bin_width = 0.1  # Width of each bin

for model in models:
    correctness_col = f'{model}_correctness'
    confidence_col = f'{model}_confidence-num'
    
    # Filter the dataset to exclude rows where confidence is 'not_mentioned' or 'cell_empty'
    df_filtered = df_merged_2[(df_merged_2[confidence_col] != 'not_mentioned') & (df_merged_2[confidence_col] != 'cell_empty')].copy()
    
    # Convert confidence to numeric if needed (replacing invalid ones with NaN or zero)
    df_filtered[confidence_col] = pd.to_numeric(df_filtered[confidence_col], errors='coerce').fillna(0)
    df_filtered[correctness_col] = pd.to_numeric(df_filtered[correctness_col], errors='coerce').fillna(0)

    # Create bins for the 'human_r' column
    df_filtered['human_bin'] = pd.cut(df_filtered['human_r'], bins=pd.interval_range(start=0, end=1, freq=bin_width))
    
    # Group by the new bins
    grouped = df_filtered.groupby('human_bin').agg(
        accuracy=(correctness_col, 'mean'),  # Assuming correctness is binary (0 or 1)
        avg_confidence=(confidence_col, 'mean')
    ).dropna()

    # Apply rolling mean for smoothing
    grouped['accuracy_smooth'] = grouped['accuracy'].rolling(window=window_size, min_periods=1).mean()
    grouped['confidence_smooth'] = grouped['avg_confidence'].rolling(window=window_size, min_periods=1).mean()
    
    # Center of the bins for plotting (scaled to 0-100 for the x-axis)
    bin_centers = [(interval.left + interval.right) / 2 * 100 for interval in grouped.index]
    
    # Create a dual y-axis plot for each model with smoothed trends
    fig, ax1 = plt.subplots()
    ax1.set_facecolor('#f8f8f8')
    # Plot smoothed accuracy on the primary y-axis (left)
    ax1.set_xlabel('Percent of test takers answered correctly (%)',fontsize=14)
    ax1.set_ylabel('Smoothed Model Accuracy (%)', color='tab:blue',fontsize=13,fontweight="bold")
    ax1.plot(bin_centers, grouped['accuracy_smooth'] * 100, color='tab:blue', label='Accuracy (Smoothed)')
    ax1.tick_params(axis='y', labelcolor='tab:blue',labelsize=13)
    ax1.set_ylim(0, 100)  # Set accuracy range from 0 to 100
    ax1.invert_xaxis()  # Invert the x-axis to start at 100
    
    # Create a second y-axis for smoothed confidence
    ax2 = ax1.twinx()
    ax2.set_ylabel('Smoothed Model Confidence', color='tab:orange',fontsize=13,fontweight="bold")
    ax2.plot(bin_centers, grouped['confidence_smooth'], color='tab:orange', label='Confidence (Smoothed)', linestyle='dashed')
    ax2.tick_params(axis='y',labelsize=13, labelcolor='tab:orange')
    ax2.set_ylim(0, 10)  # Set confidence range from 0 to 10
    
    # Add titles and grid
    plt.title(f'{name_mapping[model]}',fontsize=17)
    fig.tight_layout()
    plt.grid(True)
    plt.savefig(f'trend/{name_mapping[model]}.png', dpi=300)  
    plt.savefig(f'trend/{name_mapping[model]}.pdf', dpi=300)  
    plt.savefig(f'trend/{name_mapping[model]}.jpg', dpi=300)  
    # Show the plot for this model
    plt.show()
```

```python
from PIL import Image
import os

def stack_images_horizontally(image_folder, output_path, spacing=10):
    # Get all PNG files in the folder
    image_paths = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.lower().endswith('.png')]
    
    # Sort the image paths to ensure consistent ordering
    image_paths.sort()
    
    # Open all images
    images = [Image.open(path) for path in image_paths]
    
    # Calculate the total width and maximum height
    total_width = sum(img.width for img in images) + spacing * (len(images) - 1)
    max_height = max(img.height for img in images)
    
    # Create a new image with the calculated dimensions
    new_image = Image.new('RGB', (total_width, max_height), color='white')
    
    # Paste images horizontally
    x_offset = 0
    for img in images:
        new_image.paste(img, (x_offset, (max_height - img.height) // 2))
        x_offset += img.width + spacing
    
    # Save the result
    new_image.save(output_path, 'PNG', dpi=(300, 300))  # Set DPI to 300 for high resolution

    return len(images)  # Return the number of images processed

# Example usage
listt = ['trend_bad','trend_good']
for item in listt:
    image_folder = item
    output_path = f'./{image_folder}.png'

    num_images = stack_images_horizontally(image_folder, output_path)
    print(f"Stacked {num_images} images. Result saved to {output_path}")
```
